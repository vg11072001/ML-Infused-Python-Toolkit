```python
class Solution:
    def isPalindrome(self, s: str) -> bool:
        """
        Given a string, determine if it is a palindrome, considering only alphanumeric characters and ignoring case.

        >> using two pointers without inbuilt functions, ASCII chart and value using ord('A'), and proper two pointer technique
        #! time complexity of this code is O(n)
        -----------------------------------------------------------------
        Given an string we need to check for palindrome by using two pointers (l, r)
        
        -----------------------------------------------------------------
        """
        l, r= 0, len(s)-1
        def alphaNumeric(c):
            return ((ord('A')<= ord(c)<= ord('Z'))
                or (ord('0')<= ord(c)<= ord('9'))
                or (ord('a')<= ord(c)<= ord('z')))
        
        while l<r:
            while l<r and not alphaNumeric(s[l]):
                l+=1
            while r>l and not alphaNumeric(s[r]):
                r-=1
            if s[l].lower()!=s[r].lower():
                return False
            l +=1
            r -=1
        return True
```

```python
class Solution:
    def twoSum(self, numbers: List[int], target: int) -> List[int]:
        """
        Given an array of integers and a target value, find two numbers that add up to the target.
        You may assume that each input would have exactly one solution, and you may not use the same element twice.
        
        >> Standard Two pointer question
        #! time complexity of this code is O(n)
        -----------------------------------------------------------------
        
        -----------------------------------------------------------------
        """

        l=0
        r=len(numbers)-1

        while l<r:
            Atarget  = numbers[l] + numbers[r]

            if Atarget > target:
                r-=1
            elif Atarget < target:
                l+=1
            else:
                return [l+1, r+1]

        return []
```

```python
class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        """"
        Given an array of integers, find all unique triplets that sum up to zero.
        You may not use the same element twice.
        
        >> condition implement and for loop with 2 pointers combination
        #! time complexity of this code is O(N^2)
        -----------------------------------------------------------------
        First move 'i' arround list and the apply two pointers movement according to conditions like 2 sum is used. But 2 more conditions to add to avoid duplicates: 
        1. moving i shouldn't be in duplicate so skip (contiue) for loop
        2. moving of 'l' pointer when we append one of result, 
        3. no need to check for r bcoz l will take care the combinations is different 
        -----------------------------------------------------------------
        
        """
        nums.sort()
        res= []
        for i in range(len(nums)):
            # 1 for duplicate
            if i>0 and nums[i] ==nums[i-1]:
                continue
            # 2 for threesum conditions
            l = i+1
            r = len(nums) - 1
            while l<r:
                A_target = nums[l]+nums[r]+nums[i]
                if A_target < 0:
                    l+=1
                elif A_target > 0:
                    r-=1
                elif A_target == 0:
                    res.append([nums[i], nums[l], nums[r]])
                    # 3 for duplicate
                    l+=1
                    while l<r and nums[l] ==nums[l-1]:
                        l+=1
        return res
```

```python
class Solution:
    def areSentencesSimilar(self, sentence1: str, sentence2: str) -> bool:
        """"
        Given a sentence and another sentence, determine if you can transform one sentence into the other by only one swap of two different words.
        
        >> condition implement and with 2 pointers combination
        #! time complexity of this code is O(n)
        ----------------------------------------------------------------
        conditions:
        1. Ensuring shorter list: We make sure that s1L is the shorter sentence so that we can use it as the baseline for comparison.
        2. Prefix matching: The l1 and l2 pointers increment as long as words match from the start.
        3. Suffix matching: The r1 and r2 pointers decrement as long as words match from the end.
        4. Final check: The condition l1 > r1 ensures that the entire shorter sentence is either covered by the matched prefix or suffix. 
            This guarantees that the shorter sentence is not a middle subsequence but a valid match either as a prefix or suffix.
        ----------------------------------------------------------------
        
        """
        
        
        s1L = sentence1.split(' ')
        s2L = sentence2.split(' ')
        # s1 will be smaller one
        if len(s2L)<len(s1L):
            s1L, s2L = s2L, s1L

        l1 = 0
        l2 = 0
        # s1 has prefix common
        while l1<len(s1L) and l2<len(s2L) and s1L[l1] == s2L[l2]:
            l1, l2 = l1 +1, l2+1

        # s1 has suffux common
        r1, r2 = len(s1L)-1, len(s2L)-1
        while r1>=0 and r2>=0 and s1L[r1] == s2L[r2]:
            r1, r2 = r1-1, r2-1

        return l1>r1
```
