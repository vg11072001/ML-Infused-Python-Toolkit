```python
'''
#! 01-10
1497. Check If Array Pairs Are Divisible by k
>>> Array , HashTable, Counting

'''

class Solution:
    def canArrange(self, arr: List[int], k: int) -> bool:
            freqstore = [0]*k
            # freq store created after taking modulus with k
            for i in arr:
                remain = abs(i%k)
                freqstore[remain] +=1 # values of remain got sorted by default

            # 1st part reject:
            # when first value  thats zero as remaninder then the pairs should be dealt
            # is not divisble by 2
            if freqstore[0]%2 != 0:
                return False

            # 2nd part rejection for remaning array:
            # when 1 to k//2 (meanse first half and second half of freq) doesn't match
            # as remainder sum should be equal to k so pair number needs to be match
            for i in range(1, k//2+1):
                if freqstore[i] != freqstore[k-i]:
                    return False

            return True
```

```python
'''
#! 02-10
1331. Rank Transform of an Array

>>> Array, Hash Table, Sorting
'''
# My Solution
# O(NLogN)
class Solution:
    def arrayRankTransform(self, arr: List[int]) -> List[int]:
        
        zipped_array = list(enumerate(arr))
        sorted_array = sorted(zipped_array, key=lambda x: x[1])
        rank = [0]*len(arr)
        r=0
        hashmap = {}
        for i, num in sorted_array:
            if num in hashmap:
                rank[i]=hashmap[num]
            else:
                r+=1
                rank[i] = r
            # print(hashmap)
            hashmap[num] = rank[i]

        return rank


# Better way to write
# 1st to sort and create hashmap on a loop
# on second loop create the rank array on index basis from arr 
class Solution:
    def arrayRankTransform(self, arr: List[int]) -> List[int]:
        if not arr:
            return []
        
        # Step 1: Create a sorted copy of the array
        sorted_arr = sorted(arr)
        
        # Step 2: Create a dictionary to assign ranks
        rank_map = {}
        rank = 1
        
        # Step 3: Assign ranks to sorted elements
        for num in sorted_arr:
            if num not in rank_map:
                rank_map[num] = rank
                rank += 1
        
        # Step 4: Replace each element in the original array with its rank
        return [rank_map[num] for num in arr]
```

```python
'''
#! 04-10
2491. Divide Players Into Teams of Equal Skill

>>> Array, Hash Table, Two Pointers, Sorting
'''

# Main logic was to sort the array and 
# check for sum of initial index and from last index value == sum_each(which is required)
class Solution:
    def dividePlayers(self, skill: List[int]) -> int:
        if len(skill)%2==0:
            length = len(skill)//2
            if sum(skill)%length==0:
                sum_each = sum(skill)//length
                chemistry = 0
                skill.sort()
                for i in range(length):
                    if skill[i]+skill[len(skill)-i-1]== sum_each:
                        chemistry += skill[i] * skill[len(skill)-i-1]
                    else:
                        return -1
                
                return chemistry
            else:
                return -1
        else:
            return -1
```

```python
'''
#! 05-10
567. Permutation in String

>>> String, Hash Table, Two Pointers, Sliding Window

'''
# using Counter compare hashmap of s1 full and short string from s2
class Solution:
    def checkInclusion(self, s1: str, s2: str) -> bool:
        s1C = Counter(s1)

        for i in range(len(s2)):
            if i+len(s1)<=len(s2):
                if Counter(s2[i:i+len(s1)]) == s1C:
                    return True
        return False
```

```python
'''
#! 08-10
1963. Minimum Number of Swaps to Make the String Balanced

>>> Two Pointers, String, Stack, Greedy
'''

class Solution:
    def minSwaps(self, s: str) -> int:
        stack1 = []
        stack2 = []
        for i in s:
            if i == '[':
                stack2.append(i)
            if i == ']':
                if stack2 and stack2.pop()=='[':
                    continue
                else:
                    stack1.append(i)
                    stack2.append('[')
        return len(stack1)
```
