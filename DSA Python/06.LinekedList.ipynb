```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
class Solution:
    def splitListToParts(self, head: Optional[ListNode], k: int) -> List[Optional[ListNode]]:
        n = 0
        T1 = head
        T2 = head
        while head:
            n+=1
            head = head.next
        res = []

        if n<=k:
            while k!=0:
                if T1 is None:
                    res.append(ListNode('', None))
                    k-=1
                else:
                    res.append(ListNode(T1.val, None))
                    T1 = T1.next
                    k-=1
        else:
            size = int(n/k)
            extra = int(n%k)
            while k!=0:
                if extra >0:
                    newsize = size+1
                    extra-=1
                else:
                    newsize =size
                    extra=0
                LN= T2
                while newsize-1!=0:
                    T2 = T2.next
                    newsize-=1
                X = T2.next
                T2.next = None
                res.append(LN)
                T2 = X
                k-=1
        return res
```

```python
class Solution:
    def insertGreatestCommonDivisors(self, head: Optional[ListNode]) -> Optional[ListNode]:
        # 2807. Insert Greatest Common Divisors in Linked List
        def helper(head):
            if head.next is None:
                return 
            a = head.val
            b = head.next.val 

            if a >= b:
                if a%b==0:
                    res= b
                else:
                    while a%b !=0:
                        reminder = a%b
                        a = b
                        b = reminder
                    res= b
            else:
                if b%a==0:
                    res= a
                else:
                    while b%a !=0:
                        reminder = b%a
                        b = a
                        a = reminder
                    res= a
            next_node = head.next
            head.next = ListNode(res, next_node)
            helper(next_node)
        root = head
        helper(head)
        return root
```
