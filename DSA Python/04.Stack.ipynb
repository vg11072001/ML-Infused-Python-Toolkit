```python
class Solution:
    def isValid(self, s: str) -> bool:
        stack = []
        count = 0
        for i in s:
            if i is '(':
                stack.append(')')
            elif i is '{':
                stack.append('}')
            elif i is '[':
                stack.append(']')
            elif (stack!= []) and(stack.pop() == i):
                count+=1
        if len(s) == 2*count:
            return True
        else:
            return False
```

```python
class Solution:
    def evalRPN(self, tokens: List[str]) -> int:
        stack  = []
        # x
        for i in tokens:
            if i is '+':
                stack.append(stack.pop()+stack.pop())
            elif i is '-':
                a = stack.pop()
                b = stack.pop()
                stack.append(b-a)
            elif i is '*':
                stack.append(stack.pop() * stack.pop())
            elif i is '/':
                a = stack.pop()
                b = stack.pop()
                stack.append(int(b/a))
            else:
                stack.append(int(i))
        # print(stack)
        return stack[0]
```

```python
class Solution:
    def generateParenthesis(self, n: int) -> List[str]:
        # only add open perenthesis if open < n
        # only add a closing paranthesis if closed < open
        # valid IIF open == closed == n
        
        stack = []
        res = []

        def backtrack(OpenN, CloseN):
            if OpenN == CloseN == n:
                res.append(''.join(stack))
                return

            if OpenN <n:
                stack.append('(')
                backtrack(OpenN+1, CloseN)
                stack.pop()
            
            if CloseN <OpenN:
                stack.append(')')
                backtrack(OpenN, CloseN+1)
                stack.pop()
        backtrack(0,0)
        return res
```

```python
class Solution:
    def generateParenthesis(self, n: int) -> List[str]:
        res = []

        def dfs(openP, closeP, s):
            if openP == closeP and openP + closeP == n * 2:
                res.append(s)
                return
            
            if openP < n:
                dfs(openP + 1, closeP, s + "(")
            
            if closeP < openP:
                dfs(openP, closeP + 1, s + ")")

        dfs(0, 0, "")

        return res
```

```python
class Solution:   
    def dailyTemperatures(self, temperatures: List[int]) -> List[int]:
        """"

        1. Initialize a result array with zeros.
        2. Initialize an empty stack.
        3. Iterate through the temperatures list.
        4. For each temperature, while the stack is not empty and the temperature at the top of the stack is less than the current temperature, 
            pop the top element and calculate the difference between the current index and the popped index. 
            Update the result array at the popped index with the difference.
        5. Push the current index onto the stack.
        6. Return the result array."""
        n = len(temperatures)
        res = [0]*n
        stack =[]

        for i in range(n):
            while stack and temperatures[i]>temperatures[stack[-1]]:
                prev_index = stack.pop()
                res[prev_index]= i -prev_index
            
            stack.append(i)
        return res
```
